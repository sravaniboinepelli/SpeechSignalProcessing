# Only group delay for all languages

```
# Convert from pickle to pandas dataframe to easily train
# Imputation is not applied for possible null values

def get_dataframe_from_pickle(filename):
    with open(filename, "rb") as read_fp:
      while 1:
        try:
            df1 = pickle.load(read_fp)
            # print(df1.describe())
            # print(df1.info())
            # print(len(df1))
        except EOFError:
            print("exception end of file")
            break
    df1 = df1.fillna(0)
    return df1

# Getting only group delay pickles

punjabi_pickle = root_path+"punjab_ind/punjabi_gp.pickle"
punjabi_df = get_dataframe_from_pickle(punjabi_pickle)
print(punjabi_df)


assam_pickle = root_path+"assam_ind/assam_gp.pickle"
assam_df = get_dataframe_from_pickle(assam_pickle)
print(len(assam_df))


or_pickle = root_path+"oriya_ind/oriya_gp.pickle"
or_df = get_dataframe_from_pickle(or_pickle)
print(len(or_df))

kan_pickle = root_path+"kannada_ind/kannada_gp.pickle"
kan_df = get_dataframe_from_pickle(kan_pickle)
print(len(kan_df))

vi_pickle = root_path+"vi_ind/vi_gp.pickle"
vi_df = get_dataframe_from_pickle(vi_pickle)
print(len(vi_df))


# or_pickle =root_path+"or.pickle"
# or_df = get_dataframe_from_pickle(or_pickle)
```

```
# Concat the dataframes for all langugaes and reset indices
df = pd.concat([punjabi_df, assam_df, or_df, kan_df, vi_df]).reset_index(drop=True)


import sklearn
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split

# X represents the feature vectors and y holds the labels

X = df.iloc[:,0:-1]
print(X)
y=df['label']
print(y)
train_x, valid_x, train_y, valid_y  = train_test_split(X, y, test_size=0.20)
```

```
# Basic SVM classifier. No hyperparameter tuning or machine learning feature processing was applied

clf = svm.SVC()
clf.fit(train_x, train_y)
predict = clf.predict(valid_x)
# print(predict)
# print(valid_y)
f1 = metrics.f1_score(valid_y, predict, average='macro')
f1m = metrics.f1_score(valid_y, predict, average='micro')

accuracy = metrics.accuracy_score(valid_y, predict, normalize=True)

print("SVM, gd, accuracy: ", accuracy)
print("SVM, gd, f1 macro: ", f1)
print("SVM, gd, f1 micro: ", f1m)
```
